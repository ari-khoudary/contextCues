# null model: starting point offset only

```
def start_point(trueCongruence, bias):
  if trueCongruence == 'congruent':
    return bias
  elif trueCongruence == 'incongruent':
    return -bias
  else:
    return 0
```

```
# specify model
model = pyddm.gddm(
    starting_position = start_point,
    bound="B",
    T_dur = 4.1,
    nondecision='ndt',
    parameters={'B': (0.5, 15), 'ndt': (0.01,1.5), 'bias': (0, 1)},
    conditions = ['trueCongruence'])

pyddm.plot.model_gui_jupyter(model, data_dt=0.05, conditions={'trueCongruence':['congruent', 'incongruent', 'neutral']})
```
